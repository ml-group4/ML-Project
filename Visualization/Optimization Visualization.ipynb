## Different layersizes and structures

```python
import numpy as np
import os

rootpath = 'Optimization Results/Layerstructure/3 Layers'
results = []
structure = []
for file in os.listdir(rootpath):
    try:
        results.append(np.load(os.path.join(rootpath,file)))
        structure.append(file)
    except:
        pass
print(structure[6][0:-55])
```

```python
# 3 LAYERS different layersizes

import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4), constrained_layout=True)

#plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)
axs = axs.ravel()

axs[0].grid(True)
axs[0].plot(results[6][0], results[6][2], label=structure[6][0:-55])
axs[1].plot(results[6][0], results[6][1], label=structure[6][0:-55])
for i in range(0,3):
    axs[0].plot(results[i][0], results[i][2], label=structure[i][0:-55])
    axs[1].plot(results[i][0], results[i][1], label=structure[i][0:-55])
    
axs[0].set_xticks(np.arange(1,11))#, rotation='vertical')
axs[0].set_xlim(xmax=11,xmin=0)
#axs.set_yticks(np.arange(0,0.07,0.01))

#axs[0].set_title('Validation accuracy')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Validation accuracy')
#axs[0].legend()

axs[1].grid(True)
#axs[1].plot(results[0], results[1])
axs[1].set_xticks(np.arange(1,11))#, rotation='vertical')
axs[1].set_xlim(xmax=11,xmin=0)
#axs.set_yticks(np.arange(0,0.07,0.01))

#axs[1].set_title('Validation Loss')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Validation loss')
#axs[1].legend()
plt.legend()
[0,1,2,4,5,10,13]


plt.savefig('ValidationLoss+Accuracy_3layers.png',dpi=500)
plt.show()
```
