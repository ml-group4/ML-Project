```python
import csv
import PIL
import pathlib
from PIL import Image
from pathlib import Path
```

```python
orignial_data_name ="Training"
new_data_name = "new_Training"
wanted_img_width = 30 
wanted_img_height = 30 

orignial_data = Path(orignial_data_name)
new_data = Path(new_data_name)
```

```python
class_number =0
picture_number=0

if not orignial_data.exists():
    print("Orginaldaten Ordner nicht gefunden")


if not new_data.exists():
    print("Erzeuge Ordner für neuen Datensatz")
    new_data.mkdir()
    
    





#for all classes
for i in range (0,43):
    
    #generating folder name
    if len(str(i)) == 1:
        classfolder_name = "0000" + str(i)
    else:
        classfolder_name = "000" + str(i) 
    
    #generating folder path
    classfolder = orignial_data / classfolder_name
    new_classfolder = new_data / classfolder_name
    
    
    
    #test if folder exits
    if not classfolder.exists():
        print(classfolder_name + "not found")
    else:
        class_number=class_number+1
        
        #creating new_class folder
        if not new_classfolder.exists():
            print("Erzeuge Ordner für Klasse: " + classfolder_name)
            new_classfolder.mkdir()
        
        #loking if GT-File exists
        GT_file_name = "GT-" + classfolder_name + ".csv"
        GT_file= classfolder / GT_file_name
        if not GT_file.exists():
            print("GT-" + classfolder_name + " nicht gefunden")
        else:
            
            #loking at the pictures
            print("GT-" + classfolder_name + " gefunden, Bilder werden bearbeitet")
            with open(GT_file) as csvfile:
                reader = csv.reader(csvfile, delimiter=';')
                line_count = 0
                
                for row in reader:
                    if not line_count == 0:
                        image_path = classfolder / row[0]
                        new_image_path = new_classfolder / row[0]
                    
                        #testing if image is there
                        if not image_path.exists():
                            print("Bild " + str(row[0]) + " nicht gefunden")
                        else:
                            #processing the immage
                            picture_number=picture_number+1
                            im = Image.open(image_path)
                            new_im = im.crop((int(row[3]),int(row[4]),int(row[5]),int(row[6])))
                            new_im = new_im.resize((wanted_img_width,wanted_img_height), PIL.Image.ANTIALIAS ) 
                            new_im.save(new_image_path)
                    
                    line_count = line_count +1
        
        
print("Fertig mit " + str(class_number) + " Klassen und "+ str(picture_number) + " Bildern ")
```
